## 11 matplotlib基础

```python
import numpy as np
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

plt.plot(x, y)
```

```python
siny = y.copy()

cosy = np.cos(x)

plt.plot(x, siny)
plt.plot(x, cosy)
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red")
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle="--")
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle="--")
plt.xlim(-5, 15)
plt.ylim(0, 1)
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle="--")
plt.axis([-1, 11, -2, 2]) # x在-1和11之间，y在-2和2之间
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle="--")
plt.xlabel("x axis")
plt.ylabel("y axis")
```

```python
plt.plot(x, siny, label="sin(x)")
plt.plot(x, cosy, color="red", linestyle="--", label="cos(x)")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend() # 决定是否显示label
```
