## 03 实现简单线性回归法

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1., 2., 3., 4., 5.])
y = np.array([1., 3., 2., 3., 5.])
```

```python
plt.scatter(x, y)
plt.axis([0, 6, 0, 6])
```

```python
x_mean = np.mean(x)
y_mean = np.mean(y)
```

```python
num = 0.0 #float
d = 0.0  #float
for x_i, y_i in zip(x, y):
    num += (x_i - x_mean) * (y_i - y_mean)
    d += (x_i - x_mean) ** 2
```

```python
a = num / d
```

```python
b = y_mean - a * x_mean
```

```python
y_hat = a * x + b

plt.scatter(x, y)
plt.plot(x, y_hat, color = 'r')
plt.axis([0, 6, 0, 6])
plt.show()
```

```python
x_predict = 6
y_predict = a * x_predict + b
y_predict
```

```python
print(x.shape, y.shape)
```
