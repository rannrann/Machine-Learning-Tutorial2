## 05 衡量回归算法的标准，MSE vs MAE

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

### 波士顿房产数据

```python
import pandas as pd

# Define the URL of the downloaded dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data"

# Define the column names for the dataset (as per the dataset description)
column_names = ["CRIM", "ZN", "INDUS", "CHAS", "NOX", "RM", "AGE", "DIS", "RAD", "TAX", "PTRATIO", "B", "LSTAT", "MEDV"]

# Load the dataset into a Pandas DataFrame
boston_df = pd.read_csv(url, delim_whitespace=True, names=column_names)

# You can access the feature matrix and target variable as follows:
x = boston_df["RM"]
y = boston_df["MEDV"]

x = np.array(x)
y = np.array(y)
```

```python
x.shape
```

```python
y.shape
```

```python
plt.scatter(x, y)
```

```python
np.max(y)
```

```python
x = x[y < 50.0]
y = y[y < 50.0]
```

```python
x.shape
```

```python
y.shape
```

```python
plt.scatter(x, y)
```

### 使用简单线性回归法

```python
from model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, seed = 666)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
from SimpleLinearRegression import SimpleLinearRegression2
```

```python
reg = SimpleLinearRegression2()
reg.fit(x_train, y_train)
```

```python
reg.a_
```

```python
reg.b_
```

```python
plt.scatter(x_train, y_train)
plt.plot(x_train, reg.predict(x_train), color = 'r')
```

```python
y_predict = reg.predict(x_test)
```

```python
mse_test = np.sum((y_test - y_predict).dot(y_test - y_predict)) / len(y_test)
mse_test
```

```python
from math import sqrt

rmse_test = sqrt(mse_test)
rmse_test
```

```python
mae_test = np.sum(np.abs((y_test - y_predict))) / len(y_test)
mae_test
```
